# 10. What is the total revenue (sum of Unit Price * Quantity) generated from these sales?

# Data Visualization

```python
data['Month']=data['Date'].dt.month_name()
sns.lineplot(x=data['Month'],y=data['Total Revenue'],data=data,marker='*')
plt.title('Line Plot of Total Revenue in Months')
plt.xlabel('months')
plt.ylabel('Total revenue')
plt.xticks(rotation=90)
plt.show()
```

* the data shows the trends in total revenue of each month
* the sales is the highest in january followed by april
* decreasing trend noted with increase in time

```python
data.groupby(['Product Name'])['Units Sold'].max().sort_values(ascending=False).head(6).plot(kind='pie',autopct='%1.1f%%');
```

* the top 6 products sold are plotted
* Hanes ComfortSoft Tshirt the first with 25.6% and Forever 21 Graphic Tee the second and rest five sharing same 15.4%


```python
Which product categories generate the highest revenue?
```

```python
sns.barplot(x=data['Product Category'],y=data['Total Revenue'],data=data)
plt.title('Product categories with the highest revenue')
plt.xlabel('Product Category')
plt.ylabel('Total revenue')
plt.xticks(rotation=45)
plt.show()
```

* Electronics Category generate more revenue when compared all other categories
* Books Category fall on the 6th position 

```python
data.groupby(['Region','Product Category'])['Total Revenue'].sum().sort_values(ascending=False).plot(kind='barh');
plt.title('Regions and highest revenue producing categories')
plt.xlabel('Total revenue')
plt.show()
```

* north america has the highest sales in electronics at the same time least sales in Books
* whereas Home appliances and Beauty Products are more profitable in Europe
* Asia has shown profitable in Clothing and Sports
  

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Month', y='Total Revenue', data=data)
plt.title('Box Plot of Total Revenue by Region')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45) 
plt.show()
```

* the presence of outliers in several months suggests that there might be some months with exceptionally high or low revenue compared to general trend

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Region', y='Total Revenue', data=data)
plt.title('Violin Plot of Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)  
plt.show()
```

* the plot for Europe suggests a wider range of revenue compared to other two regions
* Asia seems to have a slightly light higher median revenue compared to North America and Europe
* the revenue distribution in North America appears relatively concentrated around the median

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Product Category', y='Units Sold', data=data)
plt.title('Line Plot of Units Sold by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

* Units sold in Clothing category is the highest followed by books and the sports
* the beauty products are the least sold
* Electronics and Home appliances has almost similar amount of units sold

```python
data['Payment Method'].value_counts().plot(kind='bar');
plt.title('Distribution of Payment Methods')
plt.xlabel('Payment Methods')
plt.show()
```

* the more frequent transactions are done with credit cards followed by debit card purchases
* purchases done with Debit cards are the least

```python
data.groupby(['Product Name'])['Total Revenue'].max().sort_values(ascending=False).head(9).plot(kind='line')
plt.title('Top revenue producing products')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)
plt.show()
```

* the top 9 revenue generating items are plotted 
* Canon EOS R5 Camera is the one generating highest revenue with Samsung Odyssey G9 Gaming Monitor at the 9th position

```python
data.groupby(['Product Category'])['Units Sold'].sum().plot(kind='bar');
plt.title('Product categories and their units sold')
plt.xlabel('Product Category')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

* Units sold in Clothing category is the highest followed by books and the sports
* the beauty products are the least sold
* Electronics and Home appliances has almost similar amount of units sold
